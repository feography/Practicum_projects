## Описание проекта

Оператор связи планирует бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, оператору нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. 

### Цель проекта 

Необходимо обучить модель на данных заказчика для прогноза оттока клиентов с метрикой качества ROC-AUC не хуже 0,85.

### Описание услуг
Оператор предоставляет два основных типа услуг: 
Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.

Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).
Также доступны такие услуги:
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

Приведем названия стобцов к змеиному регистру. Для этого напишем функцию, которая разделит слова и приведет их к нижнему регистру. колонку customerID переименнуем вручную.

```python
def snake_reg(word):
    for l in word:
        if l.isupper():
            word = word.replace(l, '_'+ l.lower())
            
    return word[1:]
```

## Подготовка данных

Выделим целевой признак и разделим данные на тестовую и обучащую выборки. Определим категориальные признаки и числовые.

```python
RANDOM_STATE = 111223
```

```python
features = data.drop('target', axis=1)
target = data['target']
```

```python
cat_columns = features.select_dtypes(include='object').columns.tolist()
num_columns = features.select_dtypes(include='number').columns.tolist()
```

```python
print(cat_columns)
print(num_columns)
```

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE, stratify=target)
```

```python
print(features_train.shape, features_test.shape)
print(target_train.shape, target_test.shape)
```

Трансформируем данные 

```python
column_transformer = make_column_transformer((StandardScaler(), num_columns),
                                             (OneHotEncoder(), cat_columns),
                                              remainder='passthrough')
column_transformer
```

## Обучение моделей

### Модель RidgeClassifier

```python
ridge = RidgeClassifier(random_state=111223, class_weight='balanced')
pipeline_ridge = make_pipeline(column_transformer, ridge) 
```

```python
pipeline_ridge.named_steps
```

```python
params_ridge = {'columntransformer__standardscaler': [StandardScaler(), MinMaxScaler()],
                'ridgeclassifier__alpha': np.arange(0.01, 0.21, 0.02)}

grid_ridge = GridSearchCV(pipeline_ridge, params_ridge, cv= 5, scoring='roc_auc')
grid_ridge.fit(features_train, target_train)
```

```python
print('Лучшие гиперпараметры модели Ridge:',grid_ridge.best_params_)
```

Для модели Ridge лучшим преобразованием численных данных оказался MinMaxScaler()

```python
print(f'Лучшее значение roc_auc модели RidgeClassifier: {grid_ridge.best_score_:.3f}')
```

### модель RandomForestClassifier

```python
rf = RandomForestClassifier(random_state=111223, class_weight='balanced')
pipeline_rf = make_pipeline(column_transformer, rf)

pipeline_rf
```

```python
pipeline_rf.named_steps
```

```python
params_rf = {'columntransformer__standardscaler': [StandardScaler(), MinMaxScaler()],
            'randomforestclassifier__max_depth': range(3, 10),
            'randomforestclassifier__min_samples_leaf': range(2, 10),
            'randomforestclassifier__n_estimators': range(10, 120, 10)}

grid_rf = RandomizedSearchCV(pipeline_rf, params_rf, cv=5, scoring='roc_auc')
grid_rf.fit(features_train, target_train)
```

```python
print('Лучшие гиперпараметры модели RandomForestClassifier:', grid_rf.best_params_)
```

```python
grid_rf.best_estimator_.named_steps['columntransformer'].transformers_[1][1].get_feature_names()
```

```python
print(f'Лучшее значение roc_auc модели RandomForestClassifier: {grid_rf.best_score_:.3f}')
```

### Модель CAtBoostClassifier

```python
#расчитаем веса классов для модели CatBoost
classes = np.unique(target_train)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=target_train)
cbc_weights = dict(zip(classes, weights))
cbc_weights
```

```python
cbc = CatBoostClassifier(random_state=111223, class_weights = cbc_weights,
                         loss_function='Logloss', verbose=False)
pipeline_cbc = make_pipeline(column_transformer, cbc)
pipeline_cbc
```

```python
pipeline_cbc.named_steps
```

```python
params_cbc = {'columntransformer__standardscaler': [StandardScaler(), MinMaxScaler()],
             'catboostclassifier__depth': range(3, 12),
             'catboostclassifier__learning_rate': np.arange(0.01, 0.25, 0.02),
             'catboostclassifier__iterations': [100, 200, 300, 400]}

grid_cbc = RandomizedSearchCV(pipeline_cbc, params_cbc, cv=5, scoring='roc_auc')
grid_cbc.fit(features_train, target_train)
```

```python
print('Лучшие гиперпараметры модели CatBoostClassifier:', grid_cbc.best_params_)
```

```python
best_cbc = grid_cbc.best_estimator_
```

```python
print(f'Лучшее значение roc_auc модели CatBoostClassifier: {grid_cbc.best_score_:.3f}')
```

Возможно стоит разработать систему скидок для клиентов с большим количеством дополнительных интернет-сервисов, тк с ростом количества этих услуг растет количество уходящих клиентов.
